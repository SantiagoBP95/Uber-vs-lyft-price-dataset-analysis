# Uber vs Lyft
En este colab, realizaremos un analisis de la información recogida de la base de datos extraida de Kaggle, la cual nos cuenta a cerca de las tarifas de Uber y Lyft en los meses de Noviembre y Diciembre en Boston MA. El propósito principal de nuestra exploracion es modelar cómo varía el precio o la tarifa en estas plataformas gracias a los factores registrados en la base de datos.

```r
rideshare <-read.csv("rideshare_kaggle.csv",head=TRUE,sep = ",")
rideshare
```

Antes de empezar con el análisis es importante realizar una limpieza a nuestra base de datos para evitar trabajar con datos duplicados o datos nulos.

```r
 newrideshare <-  na.omit(rideshare)
 newrideshare %>% distinct()

  print(paste("Fueron removidas", nrow(rideshare) - nrow(newrideshare), " filas"))
```

# **Preferencia.**
Para comenzar con nuestro análisis veremos cual de las dos compañias es más utilizada por los usuarios a la hora de solicitar un servicio de transporte desde su celular.



```r
newrideshare %>% group_by(cab_type) %>%
    summarise(count = length(id),
              '%' = (length(id) / nrow(newrideshare)) * 100)
```

Como podemos observar, de los datos registrados hay una leve tendencia del 51 % a escoger el servicio de Uber por encima de Lyft, lo anterior, partiendo del total de viajes consignados en la base.
Para verlo mejor construiremos una pequeña gráfica que nos ilustre lo visto anteriormente.


```r
newrideshare %>% ggplot(aes(x=cab_type,fill=cab_type))+
  geom_bar()+
  labs(x="Preferencia de servicio", y="Frecuencia")+
  theme_minimal()
```

# **Tiempo meteorologico**
Continueremos viendo si estas aplicacciones se ven afectadas por las condiciones climaticas a la hora de prestar el servicio al usuario.

```r
newrideshare %>% group_by(short_summary) %>%
    summarise(count = length(id),
              '%' = (length(id) / nrow(newrideshare)) * 100)
```

Observando la tabla, podemos observar que los días que más se registraron viajes, fue cuando la variable clima registra categoria "muy nublado" o "parcialmente nublado", las cuales concentran aproximadamente el 70% de los viajes realizados en las aplicaciones.
En los días con lluvia, neblina, tormenta electrica, etc. aproximadamente se  realizaron un 16% del total de viajes y el restante fueron días despejados, por ende se puede sosepchar que el clima si tiene una influencia en la demanda para estas plataformas.
Para comprender mejor esta diferencia la representaremos a traves de una gráfica

```r
newrideshare %>% ggplot(aes(short_summary, fill=short_summary)) +
    labs(x="tiempo meteorologico", title = "afectación del clima en los viajes") +
    geom_bar()+
    theme_minimal()
```

```r
newrideshare %>% ggplot(aes(x=short_summary, y=price))+
  geom_boxplot()+
  ggtitle("Boxplot del precio de tarifa segmentado por el clima")+
  labs(x="clima", y="price")+
  theme_minimal()
```

Viendo la gráfica anterior uno supondría que el tiempo meteorológico que se presenta durante el dia afectaría el precio del viaje, esto añadido a que los días que menos servicios se solicitan, son con climas lluviosos o con tormentas electricas. Lo anterior nos podría dar un indicio de que la tarifa con estas condiciones sería más costosa, sin embargo, viendo el boxplot vemos que los días despejados son en los que más varian sus tarifas y además fue el clima que tuvo la tarifa más alta.

```r
table<- data.frame(modelo2$residuals)
kable<-(table)
kable
```

**Análisis:**
Como nuestro coeficiente de correlación lineal r=0,913 > 0,5 entonces existe  una fuerte asociación entre X,Y.
Además ${r^2}$ nos indica que mediante un modelo lineal la distancia explica el 83.39% de variación del precio.
Así mismo con los resultados anteriores se puede expresar el modelo ajustado como se muestra a continuación:
 $Y= 12.1182+3.83X_i +𝜀$, donde $12.1182=β_0$ y $3.8357=β_1$. Que tiene un error típico de:  $2.018$

Resumido en la siguente tabala:


```r
fl <- aov(modelo2)
summary(fl)
```

```r
fl
```

```r
zs <- sum(modelo2$residuals**2)/55093
zs
```

```r
cme <- 224342.7/55093
cs <- 1126705.4 /cme
cs
```

Una vez realizada la regresión para el servicio Uber black, se busca determinar el grado de ajuste del modelo con respecto al CME. Para esto se desarrolla la prueba de hipótesis con una significancia del 2,5%:
$Ho:\frac{CMR}{CME}=1$ VS $H1:\frac{CMR}{CME}>1$.

Por supuesto de normalidad en la distrbución de los residuales se sabe que $\frac{CMR}{CME}$ se distribuye $F$ de paramétros 1 y $n-2$.

Por el método del $p-valor$ para prueba de hipótesis de cola derecha, considerando Ho cierta


$P(f_{1,n-2})>276690.8867)$$=2.2e-16$

Como el $P-valor$ < 0,025 se rechaza Ho y se concluye que el modelo:   $Y= 12.11.+3.83distenacia_i +𝜀$ describe mejor el comportamiento del precio respecto a la distancia.

```r
confint(modelo2,level=0.025)
```

# **Pruebas de hipotesis**


## Primera prueba de hipótesis
Como primera hipótesis se tiene: ¿La evidencia muestral es lo suficiente para argumentar que la proporción de viajes en la plataforma Uber es mayor que en Lyft?
Para ello se utilizara una 𝛂=2.5% primero veremos la propoción que tiene cada sevicio.

```r
newrideshare %>% group_by(cab_type) %>%
    summarise(count = length(id),
              '%' = (length(id) / nrow(newrideshare)) * 100)
```

$H0: \pi_u=\pi _l$ Vs $H1: \pi_u>\pi _l$

Como Nu=330568 y Nl=307408 son los suficientemente grandes entonces:        
$Pu-Pl$~$N(πu-πl,P(1-P)((\frac{1}{Nu})+(\frac{1}{Nl}) ))$

Así se puede usar una distribución normal para la prueba de hipótesis
$P=\frac{(Pu\ast Nu)+(Pl\ast Nl)}{Nu+Nl}$


Por p-valeu
se calcula sobre la evidencia muestral y suponuendo H0 cierta

$Pu-Pl$~$N(0,0.5006(1-0.5006)((\frac{1}{330568})+(\frac{1}{307408})))$

```r
prop.test(x=c(330568,307408), n=c(637976, 637976),alternative='greater', conf.level=0.025)
```

**Análisis**:con una significancia del 0.025, se rechaza Ho y se concluye que si hay una diferencia en la proporciónde viajes registrados en la base de datos por parte de Uber y Lyft, esto quiere decirque los usuarios tienen mayor preferencia por la plataforma Uber ya que el intervalo de confianza para la diferencia de proporciones no contiene a 0 y es positivo.

## Segunda prueba de hipótesis
Como segunda hipótesis se tiene:¿La evidencia muestral comprueba qué el precio promedio de la plataforma Uber es mayor al de la plataforma Lyft?

$Ho: μ_u >=\mu _l$ Vs $H1: μ_u<μl$

```r
Mu <- mean(newrideshare_uber$price)
Ml <- mean(newrideshare_lyft$price)
Mu
Ml
```

Comose analizo anteriormente el precio tiene una  gráfica asimétrica de comportamiento no normal, por lo que realizamos una transformación  Ln(U), Ln(L).
Así sus nuevas medias muestrales seria y además se utilarizara 𝛂=2.5%:

```r
lnu <- mean(log(newrideshare_uber$price))
lnl <- mean(log(newrideshare_lyft$price))
lnu
lnl
```

Como $Nu=330668>50$ y $Nl=307408>50$ se emplea el estimador S en lugar de σ por la aproximación de una distribución $T-student$ a la $Z$

$Ln(U)-Ln(L)$~$N(μLn(U)-μLn(L);\sqrt{\frac{σ^2ln(U)}{Ln(U)}+\frac{σ^2ln(L)}{Ln(L)}})$

```r
sgu <-var(log(newrideshare_uber$price))
sgl <-var(log(newrideshare_lyft$price))
sgm <-sqrt((sgu**2/lnu)+(sgl**2/lnl))
sgu
sgl
sgm
```

Usando la evidencia muestra y bajo H0 cierta $Ln(U)-Ln(L)$~$N(0;0,288)$

por el metodo de p-valeu:

```r
t.test(x=log(newrideshare_uber$price), y=log(newrideshare_lyft$price), alternative="less",conf.level=0.025)
```

```r
t.test(x=log(newrideshare_uber$price), y=log(newrideshare_lyft$price), alternative="two.sided",conf.level=0.025)$conf.int
```

**Análisis**:Con una significancia del 0.25 se rechaza Ho y teniendo en cuenta el intervalo de confianza para la diferencia de medias:[ -0.04915;-0.04906] que es negativo y no contiene a 0 se concluye que el precio promedio de la plataforma Lyft es mayor que el Uber.

```r
require(MASS)
Ajusten<-fitdistr(log(newrideshare_uber$price), "normal")
Ajusten
```

```r
Ksn<- ks.test(log(newrideshare_uber$price), "pnorm",mean=Ajusten$estimate[1],sd=Ajusten$estimate[2])
Ksn
```

## Prueba Anderson Darling

```r
#require(goftest)
Adn<- ad.test(log(newrideshare_uber$price), "pnorm",mean=Ajusten$estimate[1],sd=Ajusten$estimate[2])
Adn
```

## Prueba Shapiro-Wilk

```r
sw <- log(newrideshare_uber$price)
Swn<-shapiro.test(sw[1:5000])#como Shapiro-Wilk trabaja hasta con 5000 datos
Swn
```
